```python
import matplotlib.pyplot as plt
import pickle
import glob
import os
```

```python
for path in glob.glob('../interpolation_results/cross-domain/*.pkl'):
    file = open(path, 'rb')
    data = pickle.load(file)

    datasets = {}
    for (x, dataset), value in data.items():
        if dataset not in datasets:
            datasets[dataset] = {'x': [], 'y': []}
        datasets[dataset]['x'].append(x)
        datasets[dataset]['y'].append(value)

    # Plotting
    plt.figure(figsize=(16, 6))
    for dataset, points in datasets.items():
        if dataset == 'gsm-hard':
            plt.plot(points['x'], points['y'], label=dataset)

    plt.xlabel('Value (0 to 1)')
    plt.ylabel('Test Loss')
    plt.title(os.path.basename(path))
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
