```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
#path = '../interpolation_results/coding.pkl'
#path = '../interpolation_results/math_coding2.pkl'
#path = '../interpolation_results/bio_qa.pkl'

#path =  '../interpolation_results_final/tinycodes-orca-gsm-hard-step-interpolation.pkl'
#path = '../interpolation_results_final/tinycodes-orca-gsm-hard-step-interpolation-first-part.pkl'
path = '../interpolation_results_final/tinycodes-orca-gsm-hard-step-interpolation-first-100.pkl'

with open(path, 'rb') as f:
    data = pickle.load(f)

```

```python
fig, ax = plt.subplots(figsize=(10, 6), dpi=400)

# Create a color gradient using a colormap
colors = plt.cm.plasma(np.linspace(0, 1, len(data)))

for i, (key, values) in enumerate(data.items()):
    # if key > 5:
    #     continue
    print(min(values.values()))
    
    x = [k[0] for k in values.keys()]
    y = list(values.values())
    plt.plot(x, y, label=f'chkpt {key}', color=colors[i], alpha=0.8)

ax.set_facecolor('#f6f2ef')
plt.grid(color='#e0e1dd', linestyle='-', linewidth=1)

plt.xlabel('Interpolation Ratio')
plt.ylabel('CE Loss')
plt.title('Multi-line Plot with Color Gradient')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))  # Move legend outside the plot
plt.grid(True)
plt.tight_layout()  # Adjust layout to accommodate the legend
plt.show()
```
